```python
data = pd.read_csv("NSSO68.csv")
```

```python
df = data[data['state_1'] = 'TN']
```

```python
print("Dataset Information:")
print(df.columns)
print(df.head())
print(df.shape)
```

```python
missing_info = df.isnull().sum()
print("Missing Values Information:")
print(missing_info)
```

```python
tnnew = df[['state_1', 'District', 'Region', 'Sector', 'State_Region', 'Meals_At_Home', 'ricepds_v', 'Wheatpds_q', 'chicken_q', 'pulsep_q', 'wheatos_q', 'No_of_Meals_per_day']]
```

```python
tnnew['Meals_At_Home'].fillna(tnnew['Meals_At_Home'].mean(), inplace=True)
```

```python
def remove_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_threshold = Q1 - (1.5 * IQR)
    upper_threshold = Q3 + (1.5 * IQR)
    df = df[(df[column_name] >= lower_threshold) & (df[column_name] <= upper_threshold)]
    return df

outlier_columns = ['ricepds_v', 'chicken_q']
for col in outlier_columns:
    tnnew = remove_outliers(tnnew, col)
```

```python
tnnew['total_consumption'] = tnnew[['ricepds_v', 'Wheatpds_q', 'chicken_q', 'pulsep_q', 'wheatos_q']].sum(axis=1)
```

```python
def summarize_consumption(df, group_col):
    summary = df.groupby(group_col)['total_consumption'].sum().reset_index().sort_values(by='total_consumption', ascending=False)
    return summary
```

```python
district_summary = summarize_consumption(tnnew, 'District')
region_summary = summarize_consumption(tnnew, 'Region')

print("Top Consuming Districts:")
print(district_summary.head(4))
print("Region Consumption Summary:")
print(region_summary)
```
